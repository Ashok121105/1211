# Exploratory Data Analysis (EDA) - Credit Card Approval Dataset

This notebook explores the synthetic dataset generated for the Credit Card Approval Prediction System.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Load the Dataset

```python
df = pd.read_csv('../dataset/credit_card_data.csv')
df.head()
```

## 2. Basic Information

```python
df.info()
```

```python
df.describe()
```

## 3. Target Variable Distribution

```python
sns.countplot(x='Approved', data=df, palette='viridis')
plt.title('Distribution of Approved (1) vs Rejected (0)')
plt.show()
```

## 4. Feature Analysis (Numerical)

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.histplot(df['Annual_Income'], bins=30, kde=True, ax=axes[0], color='blue')
axes[0].set_title('Annual Income Distribution')

sns.boxplot(x='Approved', y='Annual_Income', data=df, ax=axes[1], palette='Set2')
axes[1].set_title('Annual Income by Approval Status')

plt.show()
```

## 5. Feature Analysis (Categorical)

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Employment_Status', hue='Approved', data=df, palette='pastel')
plt.title('Approval Status by Employment Status')
plt.show()
```

## 6. Correlation Matrix

```python
plt.figure(figsize=(12, 10))
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
